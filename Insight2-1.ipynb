```python
import pandas as pd
import matplotlib.pyplot as plt

# List of parties we want to analyze
parties = ['BSP', 'INC', 'AAAP', 'BJP', 'SP', 'AITC']

# Function to calculate the percentage of male and female candidates
def calculate_percentage(df):
    total_candidates = len(df)
    male_candidates = len(df[df["Sex"] == "M"])
    female_candidates = len(df[df["Sex"] == "F"])
    male_percentage = (male_candidates / total_candidates) * 100
    female_percentage = (female_candidates / total_candidates) * 100
    return male_percentage, female_percentage

# Initialize a dictionary to store the results
results = {}

# Process each election file and calculate the distribution for the parties
for year in [2004, 2009, 2014, 2019]:
    file_name = f"Elections_{year}.csv"
    df = pd.read_csv(file_name)
    df_party = df[df["Party"].isin(parties)]
    party_distribution = df_party.groupby("Party").apply(calculate_percentage)
    results[year] = dict(party_distribution)

# Convert the results dictionary to a DataFrame
results_df = pd.DataFrame(results)
print(results_df)
# Plotting the bar graph
# results_df.T.plot(kind="bar", stacked=True)
# plt.title("Distribution of Male and Female Candidates for Selected Parties")
# plt.xlabel("Election Year")
# plt.ylabel("Percentage")
# plt.xticks(rotation=0)
# plt.legend(["Male", "Female"], loc="upper left")
# plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

parties = ['BSP', 'INC', 'AAP', 'BJP', 'SP', 'AITC']

def calculate_female_percentage(df):
    total_candidates = len(df)
    female_candidates = len(df[df["Sex"] == "F"])
    female_percentage = (female_candidates / total_candidates) * 100
    return female_percentage

results = {}

for year in [2004, 2009, 2014, 2019]:
    file_name = f"Elections_{year}.csv"
    df = pd.read_csv(file_name)
    df_party = df[df["Party"].isin(parties)]
    female_distribution = df_party.groupby("Party").apply(calculate_female_percentage)
    results[year] = dict(female_distribution)

results_df = pd.DataFrame(results)

results_df.T.plot(kind="bar")
plt.title("Distribution of Female Candidates for Selected Parties")
plt.xlabel("Election Year")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.legend(title="Party", loc="upper left")
plt.show()
```
